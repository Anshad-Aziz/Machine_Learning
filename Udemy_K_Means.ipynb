```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
dataset.head()
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
import matplotlib.pyplot as mtp
```

```python
from sklearn.cluster import KMeans
wcss_list=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  wcss_list.append(kmeans.inertia_)
mtp.plot(range(1,11),wcss_list)
mtp.title('The Elobw method graph')
mtp.xlabel('NUmber of Cluster(k)')
mtp.ylabel("wcss_list")
mtp.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_predict=kmeans.fit_predict(x)
```

```python
mtp.scatter(x[y_predict==0,0],x[y_predict==0,1],s=100,c='blue',label='Cluster 1')
mtp.scatter(x[y_predict==1,0],x[y_predict==1,1],s=100,c='red',label='Cluster 2')
mtp.scatter(x[y_predict==2,0],x[y_predict==2,1],s=100,c='green',label='Cluster 3')
mtp.scatter(x[y_predict==3,0],x[y_predict==3,1],s=100,c='yellow',label='Cluster 4')
mtp.scatter(x[y_predict==4,0],x[y_predict==4,1],s=100,c='black',label='Cluster 5')
mtp.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='magenta',label='cluster 6')
mtp.title('Cluster of Customers')
mtp.xlabel('Annual income ')
mtp.ylabel('Spending score')
mtp.legend()
mtp.show()
```
