```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in data.columns:
  if data[i].dtype=='object':
    data[i]=l.fit_transform(data[i])
```

```python
data.shape
```

```python
x=data.drop('Mental disorders',axis=1)
y=data['Mental disorders']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=42)
```

```python
print('x_train',x_train.shape)
print('x_test',x_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
ytrain_pred=lr.predict(x_train)
mse=mean_squared_error(y_train,ytrain_pred)
r2=r2_score(y_train,ytrain_pred)
```

```python
print('the linear regression model performance for training set')
print('_________________________________________________________')
print('Mean_Squared_Error=',mse)
print('R2=',r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
ytrain_pred_rf=rf.predict(x_train)
mse=mean_squared_error(y_train,ytrain_pred_rf)
r2=r2_score(y_train,ytrain_pred_rf)
```

```python
print('the linear regression model performance for training set')
print('_________________________________________________________')
print('Mean_Squared_Error=',mse)
print('R2=',r2)
```
