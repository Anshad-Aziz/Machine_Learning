```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv('Mall_Customers.csv')
```

```python
 df.isnull().sum()
```

```python
df
```

```python
x=df.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
Kmeans=KMeans(n_clusters=5,init='k-means++',random_state=1)
```

```python
y=Kmeans.fit_predict(x)
```

```python
y
```

```python
plt.figure(figsize=(9,9))
plt.scatter(x[y==0,0],x[y==0,1],s=50,c='green',label='c1')
plt.scatter(x[y==1,0],x[y==1,1],s=50,c='red',label='c2')
plt.scatter(x[y==2,0],x[y==2,1],s=50,c='yellow',label='c3')
plt.scatter(x[y==3,0],x[y==3,1],s=50,c='black',label='c4')
plt.scatter(x[y==4,0],x[y==4,1],s=50,c='blue',label='c5')

plt.scatter(Kmeans.cluster_centers_[:,0],Kmeans.cluster_centers_[:,1],s=100,c='cyan')

plt.show()
```
